```python
import os
import json
import yaml
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import chardet
import random

# ------------------------------------------------
# 📁 Paths
# ------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Water Quality & Supply Prediction System\archive\water_dataX.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Water Quality & Supply Prediction System"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ------------------------------------------------
# 🧩 Phase 1: Safe Data Loading
# ------------------------------------------------
print("[INFO] Loading dataset safely...")

with open(DATA_PATH, 'rb') as f:
    raw_data = f.read(10000)
    result = chardet.detect(raw_data)
    encoding_used = result['encoding']

print(f"[INFO] Detected Encoding: {encoding_used}")

try:
    df = pd.read_csv(DATA_PATH, encoding=encoding_used)
except UnicodeDecodeError:
    print("[WARN] Fallback to latin1 encoding")
    df = pd.read_csv(DATA_PATH, encoding='latin1')

df.columns = [c.strip().replace(" ", "_") for c in df.columns]
df = df.fillna(df.median(numeric_only=True))
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# ------------------------------------------------
# 🌊 Phase 2: Feature Engineering
# ------------------------------------------------
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[numeric_cols])
df_scaled = pd.DataFrame(scaled, columns=numeric_cols)

df_scaled["Contamination_Index"] = (
    df_scaled.get("pH", 0.5)*0.3 +
    df_scaled.get("Turbidity", 0.5)*0.3 +
    df_scaled.get("Conductivity", 0.5)*0.2 +
    df_scaled.get("Dissolved_Oxygen", 0.5)*0.2
)
df_scaled["Supply_Health_Index"] = (
    df_scaled.get("Pressure", 0.5)*0.4 +
    df_scaled.get("FlowRate", 0.5)*0.4 +
    df_scaled.get("Reservoir_Level", 0.5)*0.2
)
target_col = "Contamination_Index"

# ------------------------------------------------
# 🧠 Phase 3: CNN-LSTM + Hybrid ALOA + PSO Optimization
# ------------------------------------------------

def create_model(filters, lstm_units, dropout, lr, input_shape):
    model = Sequential([
        Conv1D(filters, kernel_size=2, activation='relu', input_shape=input_shape),
        Dropout(dropout),
        LSTM(lstm_units, activation='tanh', return_sequences=False),
        Dense(32, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
                  loss='mse', metrics=['mae'])
    return model


def create_sequences(X, y, time_steps=5):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i + time_steps])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)


values = df_scaled[numeric_cols + ["Contamination_Index"]].values
time_steps = 5
X, y = create_sequences(values, values[:, -1], time_steps)
input_shape = (X.shape[1], X.shape[2])

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# --- Hybrid ALOA + PSO ---
def hybrid_optimize(iterations=5, ants=5):
    # Initial random population
    population = []
    for _ in range(ants):
        params = {
            "filters": random.choice([16, 32, 64]),
            "lstm_units": random.choice([32, 64, 128]),
            "dropout": random.uniform(0.1, 0.4),
            "lr": random.choice([0.001, 0.0005])
        }
        population.append(params)

    best_rmse = float("inf")
    best_params = None

    for i in range(iterations):
        print(f"\n[ALOA+PSO] Iteration {i+1}/{iterations}")
        for j, p in enumerate(population):
            model = create_model(**p, input_shape=input_shape)
            model.fit(X_train, y_train, epochs=8, batch_size=16, verbose=0,
                      validation_split=0.2,
                      callbacks=[EarlyStopping(patience=2, restore_best_weights=True)])
            preds = model.predict(X_test)
            rmse = np.sqrt(mean_squared_error(y_test, preds))

            print(f"   ➤ Ant {j+1}: RMSE={rmse:.4f}, Params={p}")
            if rmse < best_rmse:
                best_rmse = rmse
                best_params = p

        # PSO refinement step
        for p in population:
            # Move each parameter slightly towards the best one
            if random.random() < 0.5:
                p["filters"] = best_params["filters"]
            if random.random() < 0.5:
                p["lstm_units"] = best_params["lstm_units"]
            # small nudges for dropout/lr
            p["dropout"] = max(0.1, min(0.4, p["dropout"] + random.uniform(-0.05, 0.05)))
            p["lr"] = best_params["lr"]
    print("\n[INFO] ✅ Best Params found:", best_params)
    return best_params


best_params = hybrid_optimize(iterations=5, ants=5)

# Final model training with best params
final_model = create_model(**best_params, input_shape=input_shape)
history = final_model.fit(X_train, y_train, epochs=25, batch_size=16, verbose=1,
                          validation_split=0.2,
                          callbacks=[EarlyStopping(patience=5, restore_best_weights=True)])

# ------------------------------------------------
# 📊 Phase 4: Visualization & Evaluation
# ------------------------------------------------
y_pred = final_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"\n[RESULT] RMSE={rmse:.4f}, R²={r2:.4f}")

# 1️⃣ Hybrid Accuracy Graph
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Hybrid ALOA+PSO - Training vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_accuracy_graph.png"))
plt.show()

# 2️⃣ Comparison Graph
plt.figure(figsize=(8, 4))
plt.plot(y_test[:100], label='Actual', linewidth=2)
plt.plot(y_pred[:100], label='Predicted', linestyle='--')
plt.title("Hybrid ALOA+PSO - Actual vs Predicted (100 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Contamination Index")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_comparison_graph.png"))
plt.show()

# 3️⃣ Prediction Scatter Graph
plt.figure(figsize=(6, 5))
sns.scatterplot(x=y_test.flatten(), y=y_pred.flatten(), alpha=0.7)
plt.title("Hybrid ALOA+PSO - Prediction vs Actual")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_prediction_graph.png"))
plt.show()

# 4️⃣ Result Graph
plt.figure(figsize=(6, 4))
bars = plt.bar(["RMSE", "R²"], [rmse, r2], color=["#6dd5ed", "#2193b0"])
plt.title("Hybrid ALOA+PSO - Model Performance Metrics")
plt.ylabel("Value")
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, bar.get_height()+0.01,
             f"{bar.get_height():.3f}", ha='center', va='bottom')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_result_graph.png"))
plt.show()

# ------------------------------------------------
# 💾 Save Artifacts
# ------------------------------------------------
final_model.save(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_model.h5"))

with open(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

with open(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_config.yaml"), "w") as f:
    yaml.dump({
        "features": numeric_cols,
        "target": target_col,
        "model_params": best_params,
        "optimizer": "Hybrid ALOA + PSO",
        "time_steps": time_steps
    }, f)

with open(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_results.json"), "w") as f:
    json.dump({
        "RMSE": float(rmse),
        "R2_Score": float(r2),
        "Optimizer": "Hybrid ALOA + PSO",
        "Best_Params": best_params,
        "Rows": len(df),
        "Cols": len(numeric_cols)
    }, f, indent=4)

print("\n✅ [SUCCESS] Hybrid ALOA+PSO Results Generated & Saved At:")
print(OUTPUT_DIR)
```
