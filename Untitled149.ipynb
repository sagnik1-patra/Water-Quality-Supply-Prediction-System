```python
import os
import json
import yaml
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

# ------------------------------------------------
# 📁 Paths
# ------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Water Quality & Supply Prediction System\archive\water_dataX.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Water Quality & Supply Prediction System"

MODEL_PATH = os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_model.h5")
SCALER_PATH = os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_scaler.pkl")
CONFIG_PATH = os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_config.yaml")

# ------------------------------------------------
# 🧩 Load Model, Scaler & Config
# ------------------------------------------------
print("[INFO] Loading model and configuration...")

model = tf.keras.models.load_model(MODEL_PATH)

with open(SCALER_PATH, "rb") as f:
    scaler = pickle.load(f)

with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

features = config["features"]
time_steps = config.get("time_steps", 5)

# ------------------------------------------------
# 🧠 Load Dataset
# ------------------------------------------------
print("[INFO] Loading dataset for prediction...")

import chardet
with open(DATA_PATH, "rb") as f:
    raw_data = f.read(10000)
    enc = chardet.detect(raw_data)["encoding"]

try:
    df = pd.read_csv(DATA_PATH, encoding=enc)
except UnicodeDecodeError:
    df = pd.read_csv(DATA_PATH, encoding="latin1")

df.columns = [c.strip().replace(" ", "_") for c in df.columns]
df = df.fillna(df.median(numeric_only=True))

numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
df_scaled = pd.DataFrame(scaler.transform(df[numeric_cols]), columns=numeric_cols)

# ------------------------------------------------
# 🌊 Feature Engineering for Indices
# ------------------------------------------------
df_scaled["Contamination_Index"] = (
    df_scaled.get("pH", 0.5) * 0.3 +
    df_scaled.get("Turbidity", 0.5) * 0.3 +
    df_scaled.get("Conductivity", 0.5) * 0.2 +
    df_scaled.get("Dissolved_Oxygen", 0.5) * 0.2
)
df_scaled["Supply_Health_Index"] = (
    df_scaled.get("Pressure", 0.5) * 0.4 +
    df_scaled.get("FlowRate", 0.5) * 0.4 +
    df_scaled.get("Reservoir_Level", 0.5) * 0.2
)

# ------------------------------------------------
# 🔮 Create Time-Series Sequences for Prediction
# ------------------------------------------------
def create_sequences(X, time_steps=5):
    Xs = []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i + time_steps])
    return np.array(Xs)

values = df_scaled[features + ["Contamination_Index"]].values
X_pred = create_sequences(values, time_steps)

# ------------------------------------------------
# 🚀 Generate Predictions
# ------------------------------------------------
print("[INFO] Generating predictions...")
pred_values = model.predict(X_pred)
pred_values = pred_values.flatten()

# Align predicted values with input rows
df_result = df.iloc[time_steps:].copy()
df_result["Predicted_Contamination_Index"] = pred_values

# Add qualitative safety category
def classify_quality(ci):
    if ci < 0.3:
        return "🟢 Safe"
    elif ci < 0.7:
        return "🟡 Moderate"
    else:
        return "🔴 Contaminated"

df_result["Predicted_Quality"] = df_result["Predicted_Contamination_Index"].apply(classify_quality)

# ------------------------------------------------
# 📊 Visualizations
# ------------------------------------------------
print("[INFO] Creating prediction visualization...")

plt.figure(figsize=(10, 5))
plt.plot(df_result["Predicted_Contamination_Index"], color="teal", label="Predicted CI")
plt.title("Hybrid ALOA+PSO — Predicted Contamination Index over Time")
plt.xlabel("Sample Index")
plt.ylabel("Predicted Contamination Index")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_prediction_graph.png"))
plt.show()

# Scatter plot
plt.figure(figsize=(6, 5))
sns.histplot(df_result["Predicted_Contamination_Index"], kde=True, color="dodgerblue")
plt.title("Distribution of Predicted Contamination Index")
plt.xlabel("Predicted CI")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_prediction_distribution.png"))
plt.show()

# ------------------------------------------------
# 💾 Save Outputs
# ------------------------------------------------
result_csv = os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_predictions.csv")
result_json = os.path.join(OUTPUT_DIR, "hybrid_AquaSentinel_result_summary.json")

df_result.to_csv(result_csv, index=False)

summary = {
    "Total_Samples": len(df_result),
    "Average_Predicted_CI": float(df_result["Predicted_Contamination_Index"].mean()),
    "Safe_Zones": int((df_result["Predicted_Quality"] == "🟢 Safe").sum()),
    "Moderate_Zones": int((df_result["Predicted_Quality"] == "🟡 Moderate").sum()),
    "Contaminated_Zones": int((df_result["Predicted_Quality"] == "🔴 Contaminated").sum())
}

with open(result_json, "w") as f:
    json.dump(summary, f, indent=4)

# ------------------------------------------------
# ✅ Display Summary
# ------------------------------------------------
print("\n✅ [PREDICTION COMPLETE]")
print("Results saved at:", OUTPUT_DIR)
print("\nSummary:")
for k, v in summary.items():
    print(f"  {k}: {v}")

print("\n📁 Generated Files:")
print("  • hybrid_AquaSentinel_predictions.csv")
print("  • hybrid_AquaSentinel_prediction_graph.png")
print("  • hybrid_AquaSentinel_prediction_distribution.png")
print("  • hybrid_AquaSentinel_result_summary.json")
```
